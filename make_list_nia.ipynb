### make train list text file

```python
import os 

os.getcwd()
```

```python
import os
root_dir = './nia_8_inference/'
src_dir = 'train/images'
train = os.listdir(os.path.join(root_dir,src_dir))

print(len(train))
```

```python
f = open(root_dir+"/train_id.txt",'w')
for i in range(len(train)) :
    f.write(train[i].split(".")[0]+"\n")
f.close()
```

### make test list text file

```python
root_dir = './nia_8_inference/test_init_data'
src_dir = '/test/images'
test = os.listdir(os.path.join(root_dir + src_dir))
print(len(test))
```

```python
edges = os.listdir(os.path.join(root_dir + '/test/edges'))
labels = os.listdir(os.path.join(root_dir + '/test/labels'))
```

```python
for i in range(len(edges)):
    if edges[i] != labels[i] and edges[i] !=test[i]:
        print('e')
    print(edges[i], labels[i], test[i])
```

```python
f = open(root_dir+"/test_id.txt",'w')
for i in range(len(test)) :
    f.write(test[i].split(".")[0]+"\n")
f.close()
```
