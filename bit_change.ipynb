```python
import json
import os
import cv2
import numpy as np
from PIL import Image
from ipypb import ipb

train_root_dir = './NIA_8/train/labels'
test_root_dir = './NIA_8/test/labels'

print(os.getcwd())
train_name_ls = []
test_name_ls = []

train_labels_ls=[]
test_labels_ls=[]
for root, dirs, files in os.walk(train_root_dir):
    for file in files:
        train_name_ls.append(file[:-14])

        if file.endswith('png') and not file.endswith('scaled.png'): 
            train_path = os.path.join(root, file)
            train_labels_ls.append(train_path)

for root, dirs, files in os.walk(test_root_dir):
    for file in files:
        test_name_ls.append(file[:-14])
        if file.endswith('png') and not file.endswith('scaled.png'): 
            test_path = os.path.join(root, file)
            test_labels_ls.append(test_path)            
print('train_num', len(train_labels_ls))
print('test_num', len(test_labels_ls))
```

```python
import cv2
import numpy as np
save_train_dir = './changed_bit_image_train'
if not os.path.exists(save_train_dir):
    os.makedirs(save_train_dir)

save_test_dir = './changed_bit_image_test'
if not os.path.exists(save_test_dir):
    os.makedirs(save_test_dir)

for i in ipb(range(len(train_labels_ls))):
    img = cv2.imread(train_labels_ls[i],-1)
    img = img.astype(np.uint8)
    cv2.imwrite(save_train_dir+'/'+train_name_ls[i]+'.grayscale.png',img)

for i in ipb(range(len(test_labels_ls))):
    img = cv2.imread(test_labels_ls[i],-1)
    img = img.astype(np.uint8)
    cv2.imwrite(save_test_dir+'/'+test_name_ls[i]+'.grayscale.png',img)
```
