```python
import os
import json 
from PIL import Image
import numpy as np
import cv2
import json
```

```python
root_dir = './NIA_8/'

img_ls = []
json_ls = []
for root, dirs, files in os.walk(root_dir):
    for file in files:

        if not file.endswith('grayscale.png')and file.endswith('png') and not file.endswith('scaled.png'): 
            img_path = os.path.join(root, file)
            img_ls.append(img_path)
        if file.endswith('.json'):
            json_path = os.path.join(root, file)
            json_ls.append(json_path)

print(len(img_ls))
print(len(json_ls))
            
```

```python
image_folder = './images/'
label_folder = './labels/'
train_num = 0
test_num = 0
list = [i for i in range(len(json_ls))]
save_image_train = './train_re/images_re/'
save_label_train = './train_re/labels_re/'
save_image_test = './test_re/images_re/'
save_label_test = './test_re/labels_re/'
folder_list=[save_image_train, save_label_train, save_image_test, save_label_test]
for fname in folder_list:
    if not os.path.exists(fname):
            os.makedirs(fname)
# if not os.path.exists(save_label_train):
#         os.makedirs(save_label_train)  
for i in range(len(json_ls)):
    if i%60==0:
        print('crop and resize ... {}/{}'.format(i,len(list)))
    with open(json_ls[i], 'r', encoding='utf-8') as f:
        data = json.load(f)

    im = Image.open(json_ls[i][:-4]+'png')
    gray = Image.open(json_ls[i][:-4] + 'grayscale.png')

    x = int(data[0]['box']['x'])
    y = int(data[0]['box']['y'])
    w = int(data[0]['box']['w'])
    h = int(data[0]['box']['h'])

    margin = 0.2
    margin2 = 0.7
    #crop_img = im.crop((x, y, x+w, y+h))
    #crop_img2 = gray.crop((x, y, x+w, y+h))

    crop_img = im.crop((x-margin*w, y-margin2*h, x+w+margin*w, y+h+margin*h))
    crop_img2 = gray.crop((x-margin*w, y-margin2*h, x+w+margin*w, y+h+margin*h))

    crop_img = crop_img.resize((512, 512),Image.NEAREST)
    crop_img2 = crop_img2.resize((512, 512),Image.NEAREST)
  
    #   train/test  .
#     crop_img.save(image_folder + img_ls[i].split('/')[-1])
#     crop_img2.save(label_folder + img_ls[i].split('/')[-1][:-4]+'.grayscale.png')
#     print(i,'',img_ls[i].split('/')[-1])
    
    if i<600:
        crop_img.save(save_image_train + str(list[i]) + '.png')
        crop_img2.save(save_label_train + str(list[i]) + '.grayscale.png')
        train_num+=1
    else:
        crop_img.save(save_image_test + str(list[i]) + '.png')
        crop_img2.save(save_label_test + str(list[i]) + '.grayscale.png')
        test_num+=1
print('train_num : {}, test_num : {}'.format(train_num, test_num))
    
```
